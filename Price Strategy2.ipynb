```python
import pandas as pd
import numpy as np
from pylab import rcParams
import seaborn as sns
from sklearn import preprocessing
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv('avocado.csv' , index_col=0)
```

```python
data.drop(columns=['Date','4046','4225','4770','Total Bags','Small Bags','Large Bags','XLarge Bags','type','year','region'], inplace=True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(data['AveragePrice'], data['Total Volume'])
ax.set_xlabel('Average Price')
ax.set_ylabel('Total Volume')
plt.show()
```

```python
data['Total Volume'] = np.log(data['Total Volume'])
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(data['AveragePrice'],data['Total Volume'])
ax.set_xlabel('Average Price')
ax.set_ylabel('Total Volume')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.distplot(data['AveragePrice'])
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.distplot(data['Total Volume'])
```

```python
X = data['AveragePrice'].values.reshape(-1,1)
y = data['Total Volume'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

```python
regressor.score(X_test, y_test)
```
